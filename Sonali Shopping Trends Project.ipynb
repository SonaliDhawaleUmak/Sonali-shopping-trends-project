# Data Visualization

# 1)Category Distribution with respect to number of Occurences

# Using Barplot

```python
plt.figure(figsize = (20, 6))
ax = df["Category"].value_counts().plot(kind = 'bar')
plt.title('Category Distribution',fontweight='bold',fontsize=28,c='red')
plt.xlabel('Employment Type', weight = "bold",fontsize = 24, labelpad = 20)
plt.ylabel('Number of Occurrences', weight = "bold",fontsize = 24, labelpad = 20);
plt.show()
```

```python
##############################################################################################################################
```

#  2)Age Distribution with respect to frequency

# Using histplot

```python
plt.figure(figsize = (20, 6))
plt.hist(df['Age'],bins = 20, edgecolor = 'k')
plt.title('Age Distribution',fontweight='bold',fontsize=28,c='red')
plt.xlabel('Age',fontweight='bold',fontsize=25)
plt.ylabel('Frequency',fontweight='bold',fontsize=25)
plt.show()
```

```python
###############################################################################################################################
```

# 3)Purchase Ammount by Gender with respect to Purchase ammount(USD)

# By Using Boxplot

```python
plt.figure(figsize = (20, 6))
sns.boxplot(x='Gender', y='Purchase Amount (USD)', data=df)
plt.title('Purchase Amount by Gender',fontweight='bold',fontsize=28,c='red')
plt.xlabel('Gender',fontweight='bold',fontsize=25)
plt.ylabel('Purchase Amount (USD)',fontweight='bold',fontsize=25)
plt.show()
```
